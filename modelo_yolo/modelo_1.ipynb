```python
import os
from pathlib import Path
import cv2

from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
```

```python
model = YOLO("best_1.pt")
```

```python
results = model("sudoku_online.png")

resultado = results[0]
imagen_anotada = resultado.plot()    

plt.figure(figsize=(8, 10))
plt.imshow(imagen_anotada[:, :, ::-1])  
plt.axis('off')
plt.title('Objetos detectados por YOLO')
plt.show()
```

```python
resultado.boxes.xyxy
```

```python
x1, y1, x2, y2 = resultado.boxes.xyxy[0]

print(x1, y1, x2, y2)
```

```python
x1, y1, x2, y2 = map(int, resultado.boxes.xyxy[0])

print(x1, y1, x2, y2)
```

```python
img_recortada = resultado.orig_img[y1:y2,x1:x2]
print(resultado.orig_img.shape)
print(img_recortada.shape)
```

```python
cv2.imwrite("imagen_recortada.png", img_recortada)
```

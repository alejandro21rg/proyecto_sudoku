```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Reshape
```

```python
df = pd.read_csv(
    "data/sudoku.csv",
    nrows=100000,
    dtype=str
)
df.head()
```

```python
def string_to_array(s):
    return np.array([int(x) for x in s])
```

```python
print(type(df["puzzle"][0]))
print(df["puzzle"][0])
print(len(df["puzzle"][0]))
```

```python
print(len(str(df["puzzle"][0])))
```

```python
X = np.array(
    df["puzzle"].apply(string_to_array).tolist()
)

Y = np.array(
    df["solution"].apply(string_to_array).tolist()
)
```

```python
print(X.shape)
print(Y.shape)
```

```python
X = X / 9.0
```

```python
Y = Y - 1
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
model = Sequential([

    Dense(
        512,
        activation="relu",
        input_shape=(81,)
    ),

    Dense(
        1024,
        activation="relu"
    ),

    Dense(
        1024,
        activation="relu"
    ),

    Dense(
        81 * 9,
        activation="softmax"
    ),

    Reshape((81, 9))

])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    X_train,
    Y_train,
    validation_split=0.1,
    epochs=20,
    batch_size=128
)
```

```python
loss, acc = model.evaluate(
    X_test,
    Y_test
)

print("Accuracy:", acc)
```

```python
model.save("sudoku_solver.keras")
```
